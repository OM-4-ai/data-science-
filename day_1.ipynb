```python
def add_numbers(a, b):
    return a + b
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
my_list = [1, 2]
middle = len(my_list) // 2
middle_value = my_list[middle]
sum = add_numbers(a, b)

print("Sum is :", sum)
if sum > middle_value:

    s = set(my_list[0:middle+1])

    print("Set is :", s)

elif sum == middle_value:

    d = {middle: middle_value}

    print("Dictionary is :", d)

else:

    t = tuple(my_list[middle+1:])

    print("Tuple is :", t)
```

```python
def make_matrix(my_list):
    while len(my_list) < 9:
        my_list.append(None)

    row1 = my_list[0:3]
    row2 = my_list[3:6]
    row3 = my_list[6:9]

    matrix = [row1, row2, row3]
    return matrix

def make_stats(my_list):
    my_tuple = tuple(my_list)

    numbers_only = []
    for item in my_list:
        if item is not None and (type(item) == int or type(item) == float):
            numbers_only.append(item)

    if numbers_only:
        total = 0
        for num in numbers_only:
            total = total + num
        my_mean = total / len(numbers_only)
    else:
        my_mean = None

    if numbers_only:
        numbers_only.sort()
        middle = len(numbers_only) // 2
        my_median = numbers_only[middle]
    else:
        my_median = None

    if numbers_only:
        count_dict = {}
        for num in numbers_only:
            if num in count_dict:
                count_dict[num] = count_dict[num] + 1
            else:
                count_dict[num] = 1

        max_count = 0
        my_mode = None
        for num, count in count_dict.items():
            if count > max_count:
                max_count = count
                my_mode = num
    else:
        my_mode = None

    stats = {
        "tuple": my_tuple,
        "mean": my_mean,
        "median": my_median,
        "mode": my_mode
    }
    return stats

def main_function(my_list):
    matrix_result = make_matrix(my_list)
    stats_result = make_stats(my_list)

    final_result = {
        "matrix_3x3": matrix_result,
        "statistics": stats_result
    }
    return final_result

print("=== My First Python Program ===")
print("Enter up to 9 numbers separated by space:")

user_input = input()
input_list = user_input.split()

final_list = []
for item in input_list:
    try:
        number = int(item)
        final_list.append(number)
    except:
        try:
            number = float(item)
            final_list.append(number)
        except:
            final_list.append(None)

result = main_function(final_list)

print("\n=== RESULT ===")
print("Matrix (3x3):")
for row in result["matrix_3x3"]:
    print(row)
print("\nStatistics:")
print(result["statistics"])
```
