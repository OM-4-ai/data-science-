```python
import numpy as np
score=np.random.randint(0,101, size=50)
print("original score:/n",score)
mean_score = np.mean(score)
median_score = np.median(score)
highest_score = np.max(score)
lowest_score = np.min(score)
std_dev = np.std(score)

print(f"Mean Score:    {mean_score:.2f}")
print(f"Median Score:  {median_score:.1f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score:  {lowest_score}")
print(f"Standard Deviation: {std_dev:.2f}")
fail_mask = score < 40
distinction_mask = score > 85

failing_scores = score[fail_mask]
distinction_scores = score[distinction_mask]

print(f"Number of failures: {len(failing_scores)} -> Scores: {failing_scores}")
print(f"Number of distinctions: {len(distinction_scores)} -> Scores: {distinction_scores}")

```

```python
#normalize
normalized_scores = (score - lowest_score) / (highest_score - lowest_score)

print("First 10 Normalized Scores:\n", normalized_scores[:10])
```

```python
# (5 subjects, 10 students per subject)
reshaped_scores = score.reshape(5, 10)

subject_averages = np.mean(reshaped_scores, axis=1)

print("Reshaped 5x10 Array:\n", reshaped_scores)
print("\nRow-wise (Subject) Averages:")
```

```python
for i, avg in enumerate(subject_averages, start=1):
    print(f"Subject {i}: {avg:.2f}")
```
